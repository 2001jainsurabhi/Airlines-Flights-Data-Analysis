### Q1: What are the airlines in the dataset, accompaied by their frequencies?

```python
data.head(5)
```

```python
# checking how many airlines are in the dataset
data['airline'].nunique()
```

```python
# showing the names of the airline in the dataset
data['airline'].unique()
```

```python
# showing all the airlines with their frequencies
data['airline'].value_counts()
```

```python
# showing alll the airlines with their number of flight in bar graph
data['airline'].value_counts(ascending=True).plot.bar()
```

```python
data['airline'].value_counts(ascending=True).plot.barh(color=['lightblue', 'yellow'])
plt.title('Airline Frequency')
plt.xlabel('Frequency')
plt.ylabel('Airline')
plt.show()
```

## Q2: Show Bar graph representing of the Departure Time and Arrival Time

```python
# showing the departure time for the flight
data['departure_time'].value_counts()
```

```python
# showing the Arrival time for the flight
data['arrival_time'].value_counts()
```

```python
# showing the departure time and arrival time for the flights with their counts
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)

plt.bar(data['departure_time'].value_counts().index , data['departure_time'].value_counts().values,color = ['r','b'] )
plt.title('Departure Time')
plt.xlabel('D.Time')
plt.ylabel('D.Frequency')

plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index , data['arrival_time'].value_counts().values,color = ['r','b'] )
plt.title('Arrival Time')
plt.xlabel('A.Time')
plt.ylabel('A.Frequency')
plt.show()
```
